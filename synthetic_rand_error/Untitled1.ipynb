```python
# what is the coverage of the correct and mutation kmers?
# need location of seed k-mer (diginorm_alignment_details.txt)
# locations of false-negative nucleotides (fn.out)
# countingtable
# and correct nucleotide (mutations.txt)
import khmer
counting = khmer.load_counting_hash("diginorm.ht")
ksize = int(counting.ksize())

mutations = {}
for line in open("mutation_details.txt"):
    lexemes = line.strip().split("\t")
    if len(lexemes) != 4:
        continue

    read_name = lexemes[0]
    pos = int(lexemes[1])
    orig = lexemes[2].upper()
    mut = lexemes[3].upper()

    if read_name not in mutations:
        mutations[read_name] = {}
    if pos not in mutations[read_name]: # only take first mutation for a site
        mutations[read_name][pos] = {"correct" : orig, "mutation" : mut}

reads = {}
for line in open("fn.out"):
    line = line.strip()
    seq_a, seq_b, mutpos, name = line.split()
    mutpos = int(mutpos)
    if name not in reads:
        index = !grep ^{name}$ -A 6 diginorm_alignment_details.txt | tail -n 1
        reads[name] = {}
        reads[name]["seed"] = int(index[0])
        reads[name]["mutations"] = []
    uncorrected_kmer = corrected_kmer = False
    inside = False
    if mutpos > (reads[name]["seed"] + ksize):
        uncorrected_kmer = seq_a[(mutpos - ksize):mutpos]
        corrected_kmer = uncorrected_kmer[:-1] + mutations[name][mutpos]["correct"]
    elif mutpos < reads[name]["seed"]:
        uncorrected_kmer = seq_a[mutpos:(mutpos + ksize)]
        corrected_kmer = mutations[name][mutpos]["correct"] + uncorrected_kmer[1:]
    else:
        inside = True
    uncorrected_kmer_coverage = corrected_kmer_coverage = -1
    if uncorrected_kmer:
        uncorrected_kmer_coverage = counting.get(uncorrected_kmer)
        corrected_kmer_coverage = counting.get(corrected_kmer)
    reads[name]["mutations"].append((mutpos, corrected_kmer_coverage, uncorrected_kmer_coverage, inside))

mutations = 0
low_coverage = 0
mutation_in_seed = 0
unknown = 0
for key, value in reads.iteritems():
    index = value["seed"]
    for (mutpos, corrected_kmer_coverage, uncorrected_kmer_coverage, inside) in value["mutations"]:
        mutations += 1
        if inside:
            mutation_in_seed += 1
        elif corrected_kmer_coverage < 5 or uncorrected_kmer_coverage < 5:
            low_coverage += 1
        else:
            unknown += 1
            print "%s mutpos %d correct cov %d mut cov %d; inside? %s; seed k-mer index %d" % \
            (key, mutpos, corrected_kmer_coverage, uncorrected_kmer_coverage, inside, index)
            
print "Number of reads: %d. Number of mutations %d" % (len(reads), mutations)
print "Number of mutations whose k-mers have low coverage: %d" % low_coverage
print "Number of mutations that occur inside a seed k-mer: %d" % mutation_in_seed
print "Unclassified errors: %d" % unknown
    
            

```
